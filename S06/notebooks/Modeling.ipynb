```python
def add_noise(x,sigma):
    return x+np.random.normal(0,sigma,x.shape)
```

# $\chi^2$ distribution

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/613cc5993b3b61897dbc9a8d722487d86a34f406)

```python
df = 10

chi_data = []
for _ in range(1000):
    z = np.random.normal(0,1,df)
    chi_data.append(np.sum(z**2))
np.mean(chi_data)
```

```python
from scipy.stats import chi2
import matplotlib.pyplot as plt

mean, var, skew, kurt = chi2.stats(df, moments='mvsk')
print(mean, var, skew, kurt )
```

```python
x = np.linspace(0,chi2.ppf(0.999, df), 100)
plt.plot(x, chi2.pdf(x, df),'k')
plt.hist(chi_data, 30, density=1);
```
